# Teachable Machine - Classification of Rock, Paper, Scissors

**Procedure:**

1. Use Teachable machine to build the dataset(rock,paper,scissors1,scissors2) and train the model.
2. Download the trained model,unzip it and save them into the `model` directory.
 
    - `keras_model.h5`
    - `labels.txt`
3. Add test datasets in the sample directory
4. create conda environment(`conda create --name teachable python=3.11.7`) named `teachable` and install the following requirements.
    - `tensorflow`, which is the 2.17.x version.(`pip install tensorflow==2.12.1`)
    - PIL (Python Image Library)(`pip install pillow`)
5. Activate virtual environment. `conda activate teachable`

## Test the model with images

Import necessary packages.

```python
#import necessary packages and libraries
from keras.models import load_model
from PIL import Image, ImageOps
import numpy as np
import matplotlib.pyplot as plt
```

Load the trained model and labels file

```python
# Disable scientific notation for clarity
np.set_printoptions(suppress=True)

# Load the model
model = load_model("model/keras_model.h5", compile=False)

# Load the labels
class_names = open("model/labels.txt", "r").readlines()

# Create the array of the right shape to feed into the keras model
# The 'length' or number of images you can put into the array is
# determined by the first position in the shape tuple, in this case 1
data = np.ndarray(shape=(1, 224, 224, 3), dtype=np.float32)
```

## Test 1  :rock class

```python
# Replace this with the path to your image
image = Image.open("samples/rock.jpg").convert("RGB")

# resizing the image to be at least 224x224 and then cropping from the center
size = (224, 224)
image = ImageOps.fit(image, size, Image.Resampling.LANCZOS)

# turn the image into a numpy array
image_array = np.asarray(image)

#display tets image
plt.imshow(image_array)
plt.axis('off') 
plt.show()

# Normalize the image
normalized_image_array = (image_array.astype(np.float32) / 127.5) - 1

# Load the image into the array
data[0] = normalized_image_array

# Predicts the model
prediction = model.predict(data)
index = np.argmax(prediction)
class_name = class_names[index]
confidence_score = prediction[0][index]

# Print prediction and confidence score
print("Class:", class_name[2:], end="")
print("Confidence Score:", confidence_score)
```
